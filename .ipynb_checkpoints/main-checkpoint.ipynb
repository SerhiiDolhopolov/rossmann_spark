```python
import glob
from pathlib import Path

import seaborn as sns
import matplotlib.pyplot as plt
from minio import Minio
from pyspark.sql import SparkSession
from pyspark.sql.functions import sum as _sum

from config import S3_ENDPOINT, S3_ACCESS_KEY, S3_SECRET_KEY
```

```python
spark = SparkSession.builder \
    .appName("App") \
    .getOrCreate()
```

```python
client = Minio(
    endpoint=S3_ENDPOINT,
    access_key=S3_ACCESS_KEY,
    secret_key=S3_SECRET_KEY,
    secure=False
)
```

```python
path_temp = Path("tmp")
path_temp.mkdir(parents=True, exist_ok=True)

objects = client.list_objects(
    "archive-transactions", 
    prefix="transactions_", 
    recursive=True
)
for obj in objects:
    if obj.object_name.endswith(".parquet"):
        local_path = path_temp / obj.object_name
        local_path.parent.mkdir(parents=True, exist_ok=True)
        client.fget_object(
            "archive-transactions",
            obj.object_name,
            str(local_path),
        )

parquet_files = glob.glob(str(path_temp / "transactions_*.parquet"))

df = spark.read.parquet(*parquet_files)
```

```python
df.show(2, False, True)
```

```python
df.printSchema()
```

```python
df.count()
```

```python
agg_df = df.groupBy("category_name") \
    .agg(
        _sum("quantity").alias("total_quantity"),
        _sum("product_price").alias("total_product_price")
    ).orderBy("total_product_price", ascending=False)
```

```python
agg_df.show(10)
```

```python
pand_df = agg_df.toPandas()

fig, ax1 = plt.subplots(figsize=(10, 6))
sns.barplot(
    data=pand_df, 
    x='category_name', 
    y='total_product_price', 
    ax=ax1, 
    color='skyblue'
)
ax1.set_ylabel('Total Product Price')

ax2 = ax1.twinx()
sns.lineplot(
    data=pand_df, 
    x='category_name', 
    y='total_quantity', 
    ax=ax2, 
    color='orange', 
    marker='o'
)
ax2.set_ylabel('Total Quantity')

for i, point in pand_df.iterrows():
    ax2.text(
        x=i, 
        y=point['total_quantity'] + 0.01 * point['total_quantity'],
        s=f"{point['total_quantity']:.0f}",
        color='black', 
        fontsize=10, 
        ha='center'
    )

plt.title('Total Product Price and Quantity by Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.subplots_adjust(top=1.2) 
plt.show()
```
